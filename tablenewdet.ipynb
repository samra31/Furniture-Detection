```python
import torch
from pathlib import Path
from PIL import Image
import cv2
from tqdm import tqdm
from ultralytics.utils import LOGGER, TryExcept, ops, plt_settings, threaded
#from utils.plots import plot_one_box
from utils.general import check_img_size, non_max_suppression, scale_coords
from models.experimental import attempt_load
from utils.datasets import LoadImages
from utils.torch_utils import select_device

def detect(model, img, device):
    img = torch.from_numpy(img).to(device)
    img = img.float()  # uint8 to fp16/32
    img /= 255.0  # 0 - 255 to 0.0 - 1.0
    if img.ndimension() == 3:
        img = img.unsqueeze(0)

    pred = model(img)[0]
    pred = non_max_suppression(pred, conf_thres=0.25, iou_thres=0.45)[0]

    return pred

def main():
    device = select_device('')
    table_weights_path = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/yolov8n_custom/weights/best.pt'
    person_weights_path = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/train13/weights/best.pt'
    img_folder = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/predict5'
    table_coords_folder = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/tabcomma2'
    person_coords_folder = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/outpers2_comma'
    output_folder = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/coldet_4'

    # Load YOLOv8 models for table and person
    table_model = attempt_load(table_weights_path, map_location=device)
    table_model.eval()

    person_model = attempt_load(person_weights_path, map_location=device)
    person_model.eval()

    # Load images
    img_paths = [img for img in Path(img_folder).rglob('*.jpg')]

    for img_path in tqdm(img_paths, desc='Processing images'):
        img = cv2.imread(str(img_path))  # BGR
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # RGB

        # Detect bounding boxes for tables and persons
        table_coords = []
        person_coords = []

        for coords_folder, coords_list, model in zip(
            [table_coords_folder, person_coords_folder],
            [table_coords, person_coords],
            [table_model, person_model]
        ):
            for txt_path in Path(coords_folder).rglob('*.txt'):
                with open(txt_path, 'r') as file:
                    for line in file:
                        coords = [float(coord) for coord in line.strip().split()]
                        coords_list.append(coords)

            pred = detect(model, img_rgb, device)

        # Check for overlapping bounding boxes
        overlapping_coords = []

        for table_coord in table_coords:
            for person_coord in person_coords:
                if (
                    table_coord[0] < person_coord[2] and
                    table_coord[2] > person_coord[0] and
                    table_coord[1] < person_coord[3] and
                    table_coord[3] > person_coord[1]
                ):
                    overlapping_coords.append(table_coord)

        # Visualize and save the output
        for coord in overlapping_coords:
            x1, y1, x2, y2 = coord
            plot_one_box([x1, y1, x2, y2], img_rgb, color=(255, 0, 0), label='Overlap')

        output_path = os.path.join(output_folder, f'{img_path.stem}_overlap.jpg')
        cv2.imwrite(output_path, img_rgb)

if __name__ == "__main__":
    main()
```

```python
import torch
from pathlib import Path
import cv2
import importlib
import sys
from tqdm import tqdm
#import torch.nn as nn
#import torch.nn.functional as F
from ultralytics.utils import LOGGER, TryExcept, ops, plt_settings, threaded
from utils.general import check_img_size, non_max_suppression, scale_coords
from models.experimental import attempt_load
from utils.datasets import LoadImages
from utils.torch_utils import select_device

def detect(model, img, device):
    img = torch.from_numpy(img).to(device)
    img = img.float()  # uint8 to fp16/32
    img /= 255.0  # 0 - 255 to 0.0 - 1.0
    if img.ndimension() == 3:
        img = img.unsqueeze(0)

    pred = model(img)[0]
    pred = non_max_suppression(pred, conf_thres=0.25, iou_thres=0.45)[0]

    return pred

def draw_rectangle(image, coordinates, color, label):
    x1, y1, x2, y2 = coordinates
    cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
    cv2.putText(image, label, (int(x1), int(y1) - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

def main():
    device = select_device('')
    table_weights_path = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/yolov8n_custom/weights/best.pt'
    person_weights_path = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/train13/weights/best.pt'
    img_folder = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/predict5'
    table_coords_folder = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/tabcomma2'
    person_coords_folder = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/outpers2_comma'
    output_folder = '/content/drive/MyDrive/PerosonTableDet/ultralytics/runs/detect/coldet_4'

    # Load YOLOv8 models for table and person
    table_model = attempt_load(table_weights_path, map_location=device)
    table_model.eval()

    person_model = attempt_load(person_weights_path, map_location=device)
    person_model.eval()

    # Load images
    img_paths = [img for img in Path(img_folder).rglob('*.jpg')]

    for img_path in tqdm(img_paths, desc='Processing images'):
        img = cv2.imread(str(img_path))  # BGR
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # RGB

        # Detect bounding boxes for tables and persons
        table_coords = []
        person_coords = []

        for coords_folder, coords_list, model in zip(
            [table_coords_folder, person_coords_folder],
            [table_coords, person_coords],
            [table_model, person_model]
        ):
            for txt_path in Path(coords_folder).rglob('*.txt'):
                with open(txt_path, 'r') as file:
                    for line in file:
                        coords = [float(coord) for coord in line.strip().split()]
                        coords_list.append(coords)

            pred = detect(model, img_rgb, device)

        # Check for overlapping bounding boxes
        overlapping_coords = []

        for table_coord in table_coords:
            for person_coord in person_coords:
                if (
                    table_coord[0] < person_coord[2] and
                    table_coord[2] > person_coord[0] and
                    table_coord[1] < person_coord[3] and
                    table_coord[3] > person_coord[1]
                ):
                    overlapping_coords.append(table_coord)

        # Visualize and save the output
        for coord in overlapping_coords:
            draw_rectangle(img_rgb, coord, color=(255, 0, 0), label='Overlap')

        output_path = os.path.join(output_folder, f'{img_path.stem}_overlap.jpg')
        cv2.imwrite(output_path, img_rgb)

if __name__ == "__main__":
    main()
```
